```
def search_shops(shops_data, criteria):
    found_shops = []
    for shop in shops_data:
        match = True
        # Check for shop name match
        if 'name' in criteria and criteria['name']:
            if criteria['name'].lower() not in shop['name'].lower():
                match = False

        # Check for location keyword match
        if match and 'location_keyword' in criteria and criteria['location_keyword']:
            if criteria['location_keyword'].lower() not in shop['location'].lower():
                match = False

        # Check for item in inventory match
        if match and 'item' in criteria and criteria['item']:
            found_item = False
            for inventory_item in shop['inventory']:
                if criteria['item'].lower() == inventory_item.lower():
                    found_item = True
                    break
            if not found_item:
                match = False

        if match:
            found_shops.append(shop)
    return found_shops

print("Defined the search_shops function.")
```

```
print('--- Testing search by name ---')
criteria1 = {'name': 'Grocery Store A'}
result1 = search_shops(shops_data, criteria1)
print(f"Searching for 'Grocery Store A': {result1}")

print('\n--- Testing search by location keyword ---')
criteria2 = {'location_keyword': 'Anytown'}
result2 = search_shops(shops_data, criteria2)
print(f"Searching for shops in 'Anytown': {result2}")

print('\n--- Testing search by item ---')
criteria3 = {'item': 'Cheese'}
result3 = search_shops(shops_data, criteria3)
print(f"Searching for shops with 'Cheese': {result3}")

print('\n--- Testing combined search ---')
criteria4 = {'location_keyword': 'Otherville', 'item': 'Apples'}
result4 = search_shops(shops_data, criteria4)
print(f"Searching for shops in 'Otherville' with 'Apples': {result4}")

print('\n--- Testing no match ---')
criteria5 = {'name': 'NonExistent Shop'}
result5 = search_shops(shops_data, criteria5)
print(f"Searching for 'NonExistent Shop': {result5}")

print('\n--- Testing case-insensitivity ---')
criteria6 = {'name': 'grocery store a', 'location_keyword': 'anytown', 'item': 'MILK'}
result6 = search_shops(shops_data, criteria6)
print(f"Searching for 'grocery store a' in 'anytown' with 'MILK': {result6}")
```

```
import re

def parse_user_query(query_string):
    search_criteria = {}
    query_lower = query_string.lower()

    # Extract location keyword
    # Regex: 'near ' followed by non-greedy match of anything, until ' with' or end of string.
    location_match = re.search(r'near\s+(.+?)(?=\s+with|\s*$)', query_lower)
    if location_match:
        search_criteria['location_keyword'] = location_match.group(1).strip()

    # Extract item
    # Regex: 'with ' followed by non-greedy match of anything, until ' near' or end of string.
    item_match = re.search(r'with\s+(.+?)(?=\s+near|\s*$)', query_lower)
    if item_match:
        search_criteria['item'] = item_match.group(1).strip()

    # If no 'location_keyword' or 'item' was found, consider the entire cleaned input as a shop name.
    if not search_criteria:
        # Remove the "shops" prefix if it exists, as it's often conversational.
        name_candidate = re.sub(r'^\s*shops\s*', '', query_string, flags=re.IGNORECASE).strip()
        if name_candidate:
            search_criteria['name'] = name_candidate

    return search_criteria


print("Starting shop search CLI. Type 'quit' to exit.")

while True:
    user_input = input("Enter search query (e.g., 'shops near Anytown with Milk', or 'quit' to exit): ")

    if user_input.lower() == 'quit':
        print("Exiting shop search CLI. Goodbye!")
        break

    search_criteria = parse_user_query(user_input)

    if not search_criteria:
        print("Could not parse your query. Please try again with keywords like 'near', 'with', or provide a shop name.")
        continue
```
